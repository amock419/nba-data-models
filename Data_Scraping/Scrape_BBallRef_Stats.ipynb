```python
import scrapy
from twisted.internet import reactor
from scrapy.crawler import CrawlerProcess
from scrapy.selector import Selector
import sys
```

```python
class BBallRefSpider(scrapy.Spider):
    name = "BBallRefSpider"
    years_to_grab = range(2010, 2021)

    def start_requests(self):
        urls = [ 'https://www.basketball-reference.com/leagues/NBA_%s_per_poss.html' %(y) for y in self.years_to_grab ]
        for i, url in enumerate(urls):
            yield scrapy.Request(url=url, callback=self.parse, meta={'year' : self.years_to_grab[i]})

    def parse(self, response):
        hxs = Selector(response)
        divs = hxs.xpath("//tr[contains(@class, 'full_table')]")
        for div in divs:
            if len(div.xpath('./td//text()').extract()) == 30 and div.xpath('./td//text()').extract()[1] != '*':
                yield {
                    'year' : response.meta['year'],
                    'name' : (div.xpath('./td//text()').extract()[0]),
                    'pos' : (div.xpath('./td//text()').extract()[1]),
                    #'age' : (div.xpath('./td//text()').extract()[2]),
                    'team' : (div.xpath('./td//text()').extract()[3]),
                    'GP' : (div.xpath('./td//text()').extract()[4]),
                    'GS' : (div.xpath('./td//text()').extract()[5])
                    #'FGA_PH' : (div.xpath('./td//text()').extract()[8]),
                    #'FGP_PH' : (div.xpath('./td//text()').extract()[9]),
                    #'3PA_PH' : (div.xpath('./td//text()').extract()[11]),
                    #'3PP_PH' : (div.xpath('./td//text()').extract()[12]),
                    #'2PA_PH' : (div.xpath('./td//text()').extract()[14]),
                    #'2PP_PH' : (div.xpath('./td//text()').extract()[15]),
                    #'FTA_PH' : (div.xpath('./td//text()').extract()[17]),
                    #'FTP_PH' : (div.xpath('./td//text()').extract()[18]),
                    #'ORB_PH' : (div.xpath('./td//text()').extract()[19]),
                    #'DRB_PH' : (div.xpath('./td//text()').extract()[20]),
                    #'TRB_PH' : (div.xpath('./td//text()').extract()[21]),
                    #'AST_PH' : (div.xpath('./td//text()').extract()[22]),
                    #'STL_PH' : (div.xpath('./td//text()').extract()[23]),
                    #'BLK_PH' : (div.xpath('./td//text()').extract()[24]),
                    #'TOV_PH' : (div.xpath('./td//text()').extract()[25]),
                    #'PF_PH' : (div.xpath('./td//text()').extract()[26]),
                    #'PTS_PH' : (div.xpath('./td//text()').extract()[27])
                    #'ORTG' : (div.xpath('./td//text()').extract()[28]),
                    #'DRTG' : (div.xpath('./td//text()').extract()[29])
                }
```

```python
bball_process = CrawlerProcess(settings={
    'FEED_FORMAT': 'json',
    'FEED_URI': 'BBallRefStats.json'
})

bball_process.crawl(BBallRefSpider)
bball_process.start()
```
