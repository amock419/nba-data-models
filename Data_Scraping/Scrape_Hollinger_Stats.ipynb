```python
import scrapy
from twisted.internet import reactor
from scrapy.crawler import CrawlerProcess
from scrapy.selector import Selector
import itertools
import sys
```

```python
class HollingerSpider(scrapy.Spider):
    name = "HollingerSpider"
    years_to_grab = range(2010, 2021)
    num_pages = 11
    
    def start_requests(self):
        urls = [ 'http://insider.espn.com/nba/hollinger/statistics/_/page/%s/year/%s/qualified/false' %(n, y) for n,y in itertools.product(range(1, self.num_pages+1), self.years_to_grab) ]
        for i, url in enumerate(urls):
            yield scrapy.Request(url=url, callback=self.parse, meta={'year' : int(url.split("year/")[1].split("/qualified")[0]) })

    def parse(self, response):
        hxs = Selector(response)
        divs = hxs.xpath("//tr[contains(@class, 'player')]")
        for div in divs:
            yield {
                'year' : response.meta['year'],
                'name' : (div.xpath('./td//text()').extract()[0-14]),
                #'team' : (div.xpath('./td//text()').extract()[1-14]),
                #'GP' : (div.xpath('./td//text()').extract()[2-14]),
                'MPG' : (div.xpath('./td//text()').extract()[3-14]),
                'TS' : (div.xpath('./td//text()').extract()[4-14]),
                'ASTR' : (div.xpath('./td//text()').extract()[5-14]),
                'TOR' : (div.xpath('./td//text()').extract()[6-14]),
                'USG' : (div.xpath('./td//text()').extract()[7-14]),
                'ORR' : (div.xpath('./td//text()').extract()[8-14]),
                'DRR' : (div.xpath('./td//text()').extract()[9-14]),
                'REBR' : (div.xpath('./td//text()').extract()[10-14]),
                'PER' : (div.xpath('./td//text()').extract()[11-14]),
                'VA' : (div.xpath('./td//text()').extract()[12-14]),
                'EWA' : (div.xpath('./td//text()').extract()[13-14])
            }
```

```python
hollinger_process = CrawlerProcess(settings={
    'FEED_FORMAT': 'json',
    'FEED_URI': 'Hollinger.json'
})

hollinger_process.crawl(HollingerSpider)
hollinger_process.start()
```
