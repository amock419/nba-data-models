```python
import scrapy
from twisted.internet import reactor
from scrapy.crawler import CrawlerProcess
from scrapy.selector import Selector
import sys
```

```python
class BBallRefAdvancedSpider(scrapy.Spider):
    name = "BBallRefAdvancedSpider"
    years_to_grab = range(2010, 2021)

    def start_requests(self):
        urls = [ 'https://www.basketball-reference.com/leagues/NBA_%s_advanced.html' %(y) for y in self.years_to_grab ]
        for i, url in enumerate(urls):
            yield scrapy.Request(url=url, callback=self.parse, meta={'year' : self.years_to_grab[i]})

    def parse(self, response):
        hxs = Selector(response)
        divs = hxs.xpath("//tr[contains(@class, 'full_table')]")
        for div in divs:
            if len(div.xpath('./td//text()').extract()) == 26:
                yield {
                    'year' : response.meta['year'],
                    'name' : (div.xpath('./td//text()').extract()[0]),
                    'FTR' : (div.xpath('./td//text()').extract()[9]),
                    'OWS' : (div.xpath('./td//text()').extract()[18]),
                    'DWS' : (div.xpath('./td//text()').extract()[19]),
                    'WS' : (div.xpath('./td//text()').extract()[20]),
                    'WS48' : (div.xpath('./td//text()').extract()[21]),
                    'OBPM' : (div.xpath('./td//text()').extract()[22]),
                    'DBPM' : (div.xpath('./td//text()').extract()[23]),
                    'BPM' : (div.xpath('./td//text()').extract()[24]),
                    'VORP' : (div.xpath('./td//text()').extract()[25])
                }
```

```python
bball_adv_process = CrawlerProcess(settings={
    'FEED_FORMAT': 'json',
    'FEED_URI': 'BBallRefAdvancedStats.json'
})

bball_adv_process.crawl(BBallRefAdvancedSpider)
bball_adv_process.start()
```
